```python
def credit_code(x):
    if (x==-1)|(x==-2):
        return 4
    elif (x==0)|(x==1):
        return 3
    elif (x==2)|(x==3):
        return 2
    elif (x==4)|(x==5):
        return 1
```

```python
y_train_code=y_train.apply(credit_code)
```

```python
y_test_code=y_test.apply(credit_code)
```

```python
base_model_name=["rf","knn","gboosting"]
base_model_instance=[RandomForestClassifier(),
                     KNeighborsClassifier(),
                     GradientBoostingClassifier()]
meta_model_name=['logistic']
meta_model_instance=LogisticRegression()

base_model_features=pd.DataFrame()
base_model_feature_train=pd.DataFrame()
base_model_feature_test=pd.DataFrame()
base_model_dfs={}
for i,j in zip(base_model_name,base_model_instance):
    base_model_dfs[i]=j.fit(x_train_scale,y_train_code)
    train_pred=base_model_dfs[i].predict_proba(x_train_scale)
    temp_train=pd.DataFrame(train_pred,columns=["%s_%i" %(i,number) for number in range(1,5)],index=x_train_scale.index)
    base_model_feature_train=pd.concat([base_model_feature_train,temp_train],axis=1,ignore_index=True)
    test_pred=base_model_dfs[i].predict_proba(x_test_scale)
    temp_test=pd.DataFrame(test_pred,columns=["%s_%i" %(i,number) for number in range(1,5)],index=x_test_scale.index)
    base_model_feature_test=pd.concat([base_model_feature_test,temp_test],axis=1,ignore_index=True)


```
